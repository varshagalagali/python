```python
from wordcloud import WordCloud
from nltk.corpus import stopwords
import matplotlib.pyplot as plt 
word_string = "India is republic nation We are proud Indians . We are Love India . Jai Hind . India India India "
stopwords = set(stopwords.words('english'))
wordcloud = WordCloud(background_color = 'white',stopwords = stopwords , min_font_size = 8)
out = wordcloud.generate(word_string)
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)
plt.imshow(out , interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = stopwords.words('english')
#print(stop_words)
sent = "India is republic nation We are proud Indians . We are Love India . India India India . Jai Hind ."
word_tokens = word_tokenize(sent)
filtered_sentence = [w for w in word_tokens if w.lower() not in stop_words]
print(word_tokens)
print(filtered_sentence)
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
sent = "makeing walked running runned"
word_tokens=word_tokenize(sent)
stemmed_vocab = []
stemobj = SnowballStemmer("english")
for i in word_tokens:
  stemmed_vocab.append(stemobj.stem(i))
print(stemmed_vocab)
```
